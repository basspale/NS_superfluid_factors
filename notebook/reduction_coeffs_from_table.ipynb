```julia
function from_table_to_coefflist(table::String)
    tab_tmp = split.(split(table, "\\\n"), "&")
    
    list_R = Array{Float64,1}[]
    for i in eachindex(tab_tmp)
        filter!(x->x!="", tab_tmp[i])
        #println(tab_tmp[i])
        push!(list_R, parse.(Float64, tab_tmp[i]))
    end
    
    coeff_regions = Array{Float64,1}[]
    mat_tmp = hcat(list_R...)
    for i in 1:size(mat_tmp)[1]
        push!(coeff_regions, mat_tmp[i,:])
        filter!(x->x!=0.0, coeff_regions[i])
    end
    
    return coeff_regions
end
```

```julia
array_Rn_AB = from_table_to_coefflist(table_Rn_AB)
```

```julia
using DelimitedFiles
```

```julia
open("Murca_reduction_nbr_AB_fit_coeffs.dat", "w") do f
    println(f, "# Taken from table 2 of Gusakov, A&A 389, 702-715 (2002)")
    println(f, "#i-th row is Region i")
end
open("Murca_reduction_nbr_AB_fit_coeffs.dat", "a") do f
    writedlm(f, array_Rn_AB, ",")
end
```
